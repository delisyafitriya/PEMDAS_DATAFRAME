```python
# NO 1 buatlah sebuah DataFrame dari data jumlah produksi sampah berdasarkan Kabupaten/Kota di Jawa Barat

import pandas as pd
# baca file excel (mengimport dari excel)
df = pd.read_excel('datajabar.xlsx', sheet_name= 'Sheet1')
df
```

```python
# NO 2 
import pandas as pd
dfkeseluruhan= pd.DataFrame(df)

# Meminta input tahun dari pengguna
tahun_input = int(input("Masukkan tahun yang ingin dicek (2015-2023): "))
hasil_sampah = 0

for i, row in df.iterrows():
    if row['Tahun'] == tahun_input:
        hasil_sampah += row['Jumlah Produksi Sampah']

# Menampilkan hasil
print(f'Produksi sampah keseluruhan pada {tahun_input}: {hasil_sampah} ton')
```

```python
# NO 3

jumlah_data_pertahun = {}
for i, row in df.iterrows():
    tahun = row['Tahun']  # Ambil tahun dari baris
    sampah = row['Jumlah Produksi Sampah']
    
    if tahun in jumlah_data_pertahun:
        jumlah_data_pertahun[tahun] += sampah
    else:
        jumlah_data_pertahun[tahun] = sampah

# Menampilkan hasil
print("Data Pertahun:")
for tahun, total in jumlah_data_pertahun.items():
    print(f"Tahun {tahun}: {total} ton")
      
```

```python
# NO 4

kab_kota_pertahun = {}
for i, row in df.iterrows():
    tahun = row['Tahun']
    kabupaten_kota = row['Kabupaten Kota']
    jumlah_produksi = row['Jumlah Produksi Sampah']
    
    if 2015 <= tahun <= 2023:
        if tahun not in kab_kota_pertahun:
            kab_kota_pertahun[tahun] = {}
        
        # Jika kabupaten/kota sudah ada dalam dictionary, tambahkan jumlah produksinya
        if kabupaten_kota in kab_kota_pertahun[tahun]:
            kab_kota_pertahun[tahun][kabupaten_kota] += jumlah_produksi
        else:
            kab_kota_pertahun[tahun][kabupaten_kota] = jumlah_produksi

# Menampilkan hasil
for tahun, kab_kota in kab_kota_pertahun.items():
    print(f"Tahun {tahun}:")
    for kabupaten_kota, total in kab_kota.items():
        print(f"  {kabupaten_kota}: {total} ton")
```
